```python
class N_Queen:
    
    #######################################################################################
    def __init__(self, size):
        # le tableau a des dimensions taille x taille
        self.size = size
        # columns[r] est un nombre c si une reine est placée à la ligne r et à la colonne c.
        # columns[r] est hors de portée si aucune reine n'est placée dans la rangée r.
        # Ainsi, une fois toutes les reines placées, elles seront à des positions
        # (columns[0], 0), (columns[1], 1), ... (columns[size - 1], size - 1)
        self.columns = []
    ########################################################################################
    
    def g_size(self):
        return self.size
    ########################################################################################
    def num_queen(self):
        return len(self.columns)
    ########################################################################################
    def placer_ligne(self, column):
        self.columns.append(column)
    ########################################################################################
    def supprimer_de_la_colonne(self):
        return self.columns.pop()
    ########################################################################################
    def Placer_reine(self, column):
        
        # index de la ligne suivante
        row = len(self.columns)
 
        # verifier la colonne 
        for queen_column in self.columns:
            if column == queen_column:
                return False
 
        # vérifier la diagonale
        for queen_row, queen_column in enumerate(self.columns):
            if queen_column - queen_row == column - row:
                return False
 
        # vérifier l'autre diagonale
        for queen_row, queen_column in enumerate(self.columns):
            if ((self.size - queen_column) - queen_row
                == (self.size - column) - row):
                return False
 
        return True
   ########################################################################################
    def display(self):
        for row in range(self.size):
            for column in range(self.size):
                if column == self.columns[row]:
                    print('♛', end=' ')
                else:
                    print('.', end=' ')
            print()
   ########################################################################################
 
def placer_reine(g_size):
    """
    Afficher un échiquier pour chaque configuration possible de placement de n reines
    sur un échiquier n x n où n = taille et imprimez le nombre de ces
    configurations.
    """
    board = N_Queen(g_size)
    number_of_solutions = placer_toutes_reines(board)
    print('Number of solutions:', number_of_solutions)
########################################################################################
def placer_toutes_reines(board):
    """
     Afficher un échiquier pour chaque configuration possible de remplissage du
     embarquez avec des reines et renvoyez le nombre de ces configurations.
     
    """
    size = board.g_size()
 
    # si la carte est pleine, afficher la solution
    if size == board.num_queen():
        board.display()
        print()
        return 1
 
    number_of_solutions = 0
    # placer la reine dans la rangée suivante

    for column in range(size):
        if board.Placer_reine(column):
            board.placer_ligne(column)
            number_of_solutions += placer_toutes_reines(board)
            board.supprimer_de_la_colonne()
 
    return number_of_solutions
########################################################################################
 
 
n = int(input('Enter n: '))
placer_reine(n)
```
